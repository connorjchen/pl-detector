```python
# Implement ML model

# TODO: add hyperparameters to model? choose different models? implement neural net for fun? live prediction webpage for demo? compare to guesslang?

import sklearn as sk
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
# from sklearn.svm import SVC
# from sklearn.neighbors import KNeighborsClassifier
# from sklearn import tree
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.cluster import KMeans
# from sklearn.datasets import load_iris, load_breast_cancer

df_final = pd.read_csv("cleaned_data_set.csv")
matrix_final = df_final.to_numpy()

x = matrix_final[:,1:]
y = matrix_final[:,0]

y[y == 'java'] = 0
y[y == 'python'] = 1
y[y == 'ocaml'] = 2
y[y == 'c'] = 3
y[y == 'c++'] = 4

y=y.astype('int')

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)


logreg_model = LogisticRegression(max_iter=1200000)
nb_model = GaussianNB()

logreg_model.fit(x_train, y_train)
nb_model.fit(x_train, y_train)

print('Logistic Regression Prediction: ', logreg_model.predict(x_test))
print('Naive Bayes Prediction: ', nb_model.predict(x_test))
print('Actual Values: ', y_test)
print("Logistic Regression testing error: %2.1f%%" % (100 *(logreg_model.predict(x_test) != y_test).mean()))
print("Naive Bayes testing error: %2.1f%%" % (100 *(nb_model.predict(x_test) != y_test).mean()))

# Attempt at printing probability of prediction for each class - idk if it works, it has strong opinions
# print(100 * nb_model.predict_proba(x_test))
# for y_pred in 100 * nb_model.predict_proba(x_test):
#     for class_prob in y_pred:
#         print("Naive Bayes testing error: %2.5f%%" % (class_prob))
```
